# **Calcium (Ca)**

**Electronic Energy Levels of Calcium (Ca)**

**Spectral analysis for an alkaline earth metal, such as calcium (Ca). Alkaline earth metals (like magnesium, calcium, etc.) have two electrons in their outermost shell, making their electronic structure more complex than hydrogen-like atoms. To approximate their energy levels, we can use quantum defect theory.**

**Quantum Defect Theory**

In multi-electron systems like calcium, the inner electrons shield the nucleus, leading to a slight modification of the energy levels compared to hydrogen-like atoms. The energy levels are given by the formula:

$
E_n = -\frac{Z_{\text{eff}}^2}{2(n - \delta_l)^2}
$

Where:
- $ Z_{\text{eff}} $ is the effective nuclear charge (the positive charge felt by the outermost electron after accounting for inner electron shielding).
- $ n $ is the principal quantum number.
- $ \delta_l $ is the quantum defect, which accounts for deviations from the hydrogen-like energy levels due to electron-electron interactions and the penetration of the electron into inner shells.

**Constants for Calcium (Ca):**
- Atomic number $ Z = 20 $
- Effective nuclear charge for the outer electron $ Z_{\text{eff}} \approx 2 $ (rough approximation)
- Quantum defect $ \delta_s = 1.7 $, $ \delta_p = 0.3 $, $ \delta_d = 0.05 $

```python
!pip install periodictable qutip -q
import numpy as np
import matplotlib.pyplot as plt
import periodictable as pt
from typing import Dict, Optional
```

```python
# Constants for Calcium (Ca)
Z_eff = 2  # Approximate effective nuclear charge for calcium
n_max = 10  # Maximum principal quantum number
quantum_defects = {'s': 1.7, 'p': 0.3, 'd': 0.05}  # Quantum defects for different orbitals

# Define the Hamiltonian for an alkaline earth metal (e.g., calcium)
def alkaline_earth_hamiltonian(Z_eff, n_max, quantum_defect):
    energies = []
    for n in range(1, n_max + 1):
        if n - quantum_defect > 0:  # Ensure we don't calculate negative n-quantum_defect
            energy_n = -Z_eff**2 / (2 * (n - quantum_defect)**2)
            energies.append(energy_n)
    return np.array(energies)

# Calculate the energy levels for different orbitals (s, p, d)
energy_levels_s = alkaline_earth_hamiltonian(Z_eff, n_max, quantum_defects['s'])
energy_levels_p = alkaline_earth_hamiltonian(Z_eff, n_max, quantum_defects['p'])
energy_levels_d = alkaline_earth_hamiltonian(Z_eff, n_max, quantum_defects['d'])

# Plot the energy levels (as lines)
plt.figure(figsize=(12, 6))  # Adjusted the figure size to make it wider

# Plot 's' orbital energy levels
for n, energy in enumerate(energy_levels_s, start=1):
    plt.hlines(energy, xmin=n - 0.25, xmax=n + 0.25, color='blue', lw=2)
    plt.text(n, energy + 0.005, f'n={n} (s)', ha='center', fontsize=10)

# Plot 'p' orbital energy levels
for n, energy in enumerate(energy_levels_p, start=1):
    plt.hlines(energy, xmin=n - 0.25, xmax=n + 0.25, color='red', lw=2)
    plt.text(n, energy + 0.005, f'n={n} (p)', ha='center', fontsize=10)

# Plot 'd' orbital energy levels
for n, energy in enumerate(energy_levels_d, start=1):
    plt.hlines(energy, xmin=n - 0.25, xmax=n + 0.25, color='green', lw=2)
    plt.text(n, energy + 0.005, f'n={n} (d)', ha='center', fontsize=10)

# Adjust the x-axis range and add space between the labels
plt.xlim(0, n_max + 2)  # Extend the x-axis a bit beyond n_max
plt.xlabel('Principal Quantum Number (n)')
plt.ylabel('Energy (a.u.)')
plt.title('Electronic Energy Levels of Calcium (Ca)')
plt.grid(True)
plt.show()

# Print energy levels
print("Energy levels (in atomic units):")
for n, (energy_s, energy_p, energy_d) in enumerate(zip(energy_levels_s, energy_levels_p, energy_levels_d), start=1):
    print(f"n={n}, s: {energy_s:.6f} a.u., p: {energy_p:.6f} a.u., d: {energy_d:.6f} a.u.")
```

1. **Quantum Defect Theory**: This model uses quantum defects to adjust the energy levels for each orbital type:
   - $ s $-orbital quantum defect $ \delta_s = 1.7 $
   - $ p $-orbital quantum defect $ \delta_p = 0.3 $
   - $ d $-orbital quantum defect $ \delta_d = 0.05 $
   
   The energy levels are adjusted accordingly for each orbital.

2. **Plotting**: The code plots the energy levels for the **s**, **p**, and **d** orbitals as separate lines with different colors.

3. **Effective Nuclear Charge**: We use an approximate effective nuclear charge $ Z_{\text{eff}} = 2 $ for calcium, which accounts for the shielding effects from inner electrons.

**Output:**

- A plot showing the **electronic energy levels** for the **s**, **p**, and **d** orbitals for calcium.
- The energy levels are printed in **atomic units** (Hartrees).

**Adapting to Other Alkaline Earth Metals:**
For other alkaline earth metals (like **magnesium**, **strontium**, etc.), you can adjust:
- The **atomic number** $ Z $
- The **effective nuclear charge** $ Z_{\text{eff}} $ based on shielding.
- The **quantum defect** values specific to the element.

This approach provides an approximation of the energy spectrum, and for more detailed results, a full quantum mechanical solution with electron-electron interactions would be necessary.

**Emission and Absorption Spectrum of Calcium (Ca)**

For the emission and absorption spectra of an atom like calcium, is different from the molecular case, since this is about electronic transitions instead of vibrational modes.

Calcium has notable spectral lines, especially the ones corresponding to transitions between the 4s, 3d, and 4p orbitals.

```python
# Step 1: Define the energy levels of calcium
# These are simplified and approximate values for major transitions in eV
energy_levels = {
    '4s': 0.0,  # Ground state (4s)
    '4p': 2.93, # First excited state (4p)
    '3d': 1.88  # Second excited state (3d)
}

# Step 2: Define possible transitions (in eV)
transitions = {
    ('4p', '4s'): energy_levels['4p'] - energy_levels['4s'],
    ('3d', '4s'): energy_levels['3d'] - energy_levels['4s'],
    ('4p', '3d'): energy_levels['4p'] - energy_levels['3d']
}

# Step 3: Convert transition energies from eV to wavelength in nm (using E = hc/λ)
def energy_to_wavelength(energy_eV):
    h = 4.1357e-15  # Planck's constant in eV·s
    c = 3e8  # Speed of light in m/s
    return (h * c) / energy_eV * 1e9  # Convert to nm

# Calculate wavelengths of transitions
wavelengths = {transition: energy_to_wavelength(energy) for transition, energy in transitions.items()}

# Step 4: Simulate and plot the emission and absorption spectra
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Emission Spectrum (peaks at specific wavelengths)
axs[0].bar(wavelengths.values(), [1.0] * len(wavelengths), width=20.0, color='orange')
axs[0].set_xlabel('Wavelength (nm)')
axs[0].set_ylabel('Intensity (a.u.)')
axs[0].set_title('Emission Spectrum of Calcium (Electronic Transitions)')
axs[0].grid(True)

# Absorption Spectrum (dips at specific wavelengths)
axs[1].plot(np.linspace(200, 1000, 1000), np.ones(1000), color='gray')  # Baseline at 1.0 intensity
for wavelength in wavelengths.values():
    axs[1].plot([wavelength, wavelength], [1.0, 0.0], color='blue', lw=2)  # Vertical absorption dips
axs[1].set_xlabel('Wavelength (nm)')
axs[1].set_ylabel('Normalized Intensity (a.u.)')
axs[1].set_title('Absorption Spectrum of Calcium (Electronic Transitions)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Step 5: Print out the transition wavelengths
print("Electronic transitions (wavelengths in nm):")
for transition, wavelength in wavelengths.items():
    print(f"{transition[0]} -> {transition[1]}: {wavelength:.2f} nm")
```

- **Energy Levels:** We are using simplified energy values for the 4s, 4p, and 3d orbitals of calcium.
- **Transitions:** We calculate the possible transitions between these states and their corresponding energies.
- **Wavelengths:** The energy of each transition is converted into a wavelength (in nm) using $ E = \frac{hc}{\lambda} $
- **Spectra:** The emission spectrum shows peaks where light is emitted, while the absorption spectrum shows dips where light is absorbed.
* Emission spectrum (peaks at specific wavelengths corresponding to transitions). Absorption spectrum (dips at the same wavelengths). Printed list of the wavelengths corresponding to the transitions.
